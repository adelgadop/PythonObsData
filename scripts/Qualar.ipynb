## O que acontece nas estações tipo urbano?

```python
urban = stations.loc[stations.type == 'Urban']
urban
```

```python
fig, ax = plt.subplots(sharex=True, figsize=(6,4))
t = 'Urban'
alpha = .1
df = obs.loc[obs.code.isin(urban.code),['o3', 'no', 'no2', 'co','tol']]
mean = df.groupby(df.index.hour).mean()
sd = df.groupby(df.index.hour).std()
    
mean.plot(y=['o3', 'no', 'no2'], style=['-o','-','-s'],
          color=['b', 'darkorange','r'], lw=3,ax=ax, legend=False)
ax.fill_between(mean.index, mean.o3+sd.o3,mean.o3-sd.o3, color='b', alpha=alpha )
ax.set_title(t, loc='left', fontsize=8)
ax.set_xlabel('Hours (Local Time)')
ax.set_ylabel('O$_3$, NO, NO$_2$ [$\mu$g m$^{-3}$]', loc='center')
ax.set_ylim(0,140)
plt.xticks(np.arange(0,24,2))
ax.legend(['O$_3$','NO','NO$_2$'], ncol=3, fontsize=8, loc=2)
    
ax2 = ax.twinx()
mean.plot(y=['co','tol'], ax=ax2, style=['-^g','-.c'], lw=2.5, legend=False)
ax2.set_ylim(0,14)
ax2.set_ylabel('CO [ppm], Toluene [$\mu$gm$^{-3}$]',loc='center')
ax2.legend(['CO','Tol'], fontsize=8, loc=1, ncol=3);
#fig.savefig('../4_Draft_Report/Dissertation/fig/byhour_all_polls.pdf', bbox_inches='tight', facecolor='w')
```
